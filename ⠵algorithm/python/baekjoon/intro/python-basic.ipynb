## ìˆ˜ ìë£Œí˜•ì˜ ì—°ì‚°
- ìˆ˜ ìë£Œí˜•ì— ëŒ€í•˜ì—¬ ì‚¬ì¹™ì—°ì‚°ê³¼ ë‚˜ë¨¸ì§€ ì—°ì‚°ìê°€ ë§ì´ ì‚¬ìš©ëœë‹¤.
- ë‹¨ ë‚˜ëˆ„ê¸° ì—°ì‚°ì(`/`)ë¥¼ ì£¼ì˜í•´ì„œ ì‚¬ìš©í•´ì•¼í•œë‹¤.
    - íŒŒì´ì¬ì—ì„œ ë‚˜ëˆ„ê¸° ì—°ì‚°ì(`/`)ëŠ” ë‚˜ëˆ ì§„ ê²°ê³¼ë¥¼ ì‹¤ìˆ˜í˜•ìœ¼ë¡œ ë°˜í™˜í•œë‹¤.
- ë‹¤ì–‘í•œ ë¡œì§ì„ ì„¤ê³„í•  ë•Œ ë‚˜ë¨¸ì§€ ì—°ì‚°ì(`%`)ë¥¼ ì´ìš©í•´ì•¼ í•  ë•Œê°€ ë§ë‹¤.
- íŒŒì´ì¬ì—ì„œëŠ” ëª«ì„ ì–»ê¸° ìœ„í•´ ëª« ì—°ì‚°ì(`//`)ë¥¼ ì‚¬ìš©í•œë‹¤.

- ì—¬ëŸ¬ ê°œì˜ ë°ì´í„°ë¥¼ ì—°ì†ì ìœ¼ë¡œ ë‹´ì•„ ì²˜ë¦¬í•˜ê¸° ìœ„í•´ ì‚¬ìš©í•˜ëŠ” ìë£Œí˜•ì´ë‹¤.

### ë¦¬ìŠ¤íŠ¸ì˜ ì¸ë±ì‹±ê³¼ ìŠ¬ë¼ì´ì‹±
- ì¸ë±ìŠ¤ ê°’ì„ ì…ë ¥í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì˜ íŠ¹ì •í•œ ì›ì†Œì— ì ‘ê·¼í•˜ëŠ” ê²ƒì„ ì¸ë±ì‹±(Indexing)ì´ë¼ê³  í•œë‹¤.
    - íŒŒì´ì¬ì˜ ì¸ë±ìŠ¤ ê°’ì€ ì–‘ì˜ ì •ìˆ˜ì™€ ìŒì˜ ì •ìˆ˜ë¥¼ ëª¨ë‘ ì‚¬ìš©í•  ìˆ˜ ìˆë‹¤.

````python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# ì—¬ëŸ ë²ˆì§¸ ì›ì†Œë§Œ ì¶œë ¥
print(a[7])

# ë’¤ì—ì„œ ì²« ë²ˆì§¸ ì›ì†Œ ì¶œë ¥
print(a[-1])

# ë’¤ì—ì„œ ì„¸ ë²ˆì§¸ ì›ì†Œ ì¶œë ¥
print(a[-3])

# ë„¤ ë²ˆì§¸ ì›ì†Œ ê°’ ë³€ê²½
a[3] = 7
print(a)
````

- ë¦¬ìŠ¤íŠ¸ì—ì„œ ì—°ì†ì ì¸ ìœ„ì¹˜ë¥¼ ê°–ëŠ” ì›ì†Œë“¤ì„ ê°€ì ¸ì™€ì•¼ í•  ë•ŒëŠ” ìŠ¬ë¼ì´ì‹±(Slicing)ì„ ì´ìš©í•œë‹¤.
    - ëŒ€ê´„í˜¸ ì•ˆì— ì½œë¡ (`:`)ì„ ë„£ì–´ì„œ **ì‹œì‘ ì¸ë±ìŠ¤** ì™€ **ë ì¸ë±ìŠ¤** ë¥¼ ì„¤ì •í•  ìˆ˜ ìˆë‹¤.

````python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# ë„¤ ë²ˆì§¸ ì›ì†Œë§Œ ì¶œë ¥
print(a[3])

# ë‘ ë²ˆì§¸ ì›ì†Œë¶€í„° ë„¤ ë²ˆì§¸ ì›ì†Œê¹Œì§€
print(a[1:4])
````

### ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜
- ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™”í•˜ëŠ” ë°©ë²• ì¤‘ í•˜ë‚˜ì´ë‹¤.
    - **ëŒ€ê´„í˜¸ ì•ˆì— ì¡°ê±´ë¬¸ê³¼ ë°˜ë³µë¬¸ì„ ì ìš©í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™” í•  ìˆ˜ ìˆë‹¤.**

````python
# 0ë¶€í„° 9ê¹Œì§€ì˜ ìˆ˜ë¥¼ í¬í•¨í•˜ëŠ” ë¦¬ìŠ¤íŠ¸
array = [i for i in range(10)]
print(array)

# 0ë¶€í„° 19ê¹Œì§€ì˜ ìˆ˜ ì¤‘ì—ì„œ í™€ìˆ˜ë§Œ í¬í•¨í•˜ëŠ” ë¦¬ìŠ¤íŠ¸
array = [i for i in range(20) if i % 2 == 1]
print(array)

# 1ë¶€í„° 9ê¹Œì§€ì˜ ìˆ˜ë“¤ì˜ ì œê³± ê°’ì„ í¬í•¨í•˜ëŠ” ë¦¬ìŠ¤íŠ¸
array = [i * i for i in range(1, 10)]
print(array)
````

- ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜ì€ 2ì°¨ì› ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™”í•  ë•Œ íš¨ê³¼ì ìœ¼ë¡œ ì‚¬ìš©ëœë‹¤.
- íŠ¹íˆ N * M í¬ê¸°ì˜ 2ì°¨ì› ë¦¬ìŠ¤íŠ¸ë¥¼ í•œ ë²ˆì— ì´ˆê¸°í™” í•´ì•¼ í•  ë•Œ ë§¤ìš° ìœ ìš©í•˜ë‹¤.
    - ì¢‹ì€ ì˜ˆì‹œ : `array = [[0] * m for _ in range(n)]`
    - ì˜ëª»ëœ ì˜ˆì‹œ : `array = [[0] * m] * n`
    â‡’ ì „ì²´ ë¦¬ìŠ¤íŠ¸ ì•ˆì— í¬í•¨ëœ ê° ë¦¬ìŠ¤íŠ¸ê°€ ëª¨ë‘ ê°™ì€ ê°ì²´ë¡œ ì¸ì‹ë˜ì–´ ê°™ì€ ì£¼ì†Œê°’ì„ ê°€ì§€ê²Œ ëœë‹¤. ì¦‰, í•˜ë‚˜ì˜ ê°’ë§Œ ë°”ë€Œì–´ë„ ë‹¤ ë°”ë€Œê²Œ ëœë‹¤.

````python
# ì¢‹ì€ ì˜ˆì‹œ
# N X M í¬ê¸°ì˜ 2ì°¨ì› ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™”
n = 4
m = 3
array = [[0] * m for _ in range(n)]
print(array)

array[1][1] = 5
print(array)
````

````python
# ì˜ëª»ëœ ì˜ˆì‹œ
# N X M í¬ê¸°ì˜ 2ì°¨ì› ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™” (ì˜ëª»ëœ ë°©ë²•)
n = 4
m = 3
array = [[0] * m] * n
print(array)

array[1][1] = 5
print(array)
````

### ë¦¬ìŠ¤íŠ¸ ê´€ë ¨ ê¸°íƒ€ ë©”ì„œë“œ

````python
a = [1, 4, 3]
print("ê¸°ë³¸ ë¦¬ìŠ¤íŠ¸ : ", a)

# ë¦¬ìŠ¤íŠ¸ì— ì›ì†Œ ì‚½ì…
a.append(2)
print("ì‚½ì… : ", a)

# ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬
a.sort()
print("ì˜¤ë¦„ì°¨ìˆœ ì •ë ¬ : ", a)

# ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
a.sort(reverse = True)
print("ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ : ", a)

# ë¦¬ìŠ¤íŠ¸ ì›ì†Œ ë’¤ì§‘ê¸°
a.reverse()
print("ì›ì†Œ ë’¤ì§‘ê¸° : ", a)

# íŠ¹ì • ì¸ë±ìŠ¤ì— ë°ì´í„° ì¶”ê°€
a.insert(2, 3)
print("ì¸ë±ìŠ¤ 2ì— 3 ì¶”ê°€ : ", a)
             
# íŠ¹ì • ê°’ì¸ ë°ì´í„° ê°œìˆ˜ ì„¸ê¸°
print("ê°’ì´ 3ì¸ ë°ì´í„° ê°œìˆ˜ : ", a.count(3))
````

````python
a = [1, 2, 3, 4, 5, 5, 5]
print("ì›ë³¸ ë¦¬ìŠ¤íŠ¸ : ", a)
# íŠ¹ì • ê°’ ë°ì´í„° ì‚­ì œ
a.remove(1)
print("ê°’ì´ 1ì¸ ë°ì´í„° ì‚­ì œ : ", a)

# ë¦¬ìŠ¤íŠ¸ì—ì„œ íŠ¹ì • ê°’ì„ ê°€ì§€ëŠ” ì›ì†Œ ëª¨ë‘ ì œê±°í•˜ê¸°
remove_set = {3, 5} # ì§‘í•© ìë£Œí˜•

#remove_setì— í¬í•¨ë˜ì§€ ì•Šì€ ê°’ë§Œì„ ì €ì¥
result = [i for i in a if i not in remove_set]
print(result)
````

## ì–¸ë”ë°”(`_`) ì‚¬ìš© ì˜ˆì‹œ
- íŒŒì´ì¬ì—ì„œëŠ” ë°˜ë³µì„ ìˆ˜í–‰í•˜ë˜ ë°˜ë³µì„ ìœ„í•œ ë³€ìˆ˜ì˜ ê°’ì„ ë¬´ì‹œí•˜ê³ ì í•  ë•Œ ì–¸ë”ë°”(`_`)ë¥¼ ìì£¼ ì‚¬ìš©í•œë‹¤.

````python
# 1ë¶€í„° 9ê¹Œì§€ì˜ ìì—°ìˆ˜ë¥¼ ë”í•˜ê¸°
summary = 0
for i in range(1, 10):
    summary += i
print(summary)

# "Hello World"ë¥¼ 5ë²ˆ ì¶œë ¥í•˜ê¸°
for _ in range(5):
    print("Hello World")
````

- `ìˆœì—´` : ì„œë¡œ ë‹¤ë¥¸ nê°œì—ì„œ ì„œë¡œ ë‹¤ë¥¸ `r`ê°œë¥¼ ì„ íƒí•˜ì—¬ ì¼ë ¬ë¡œ ë‚˜ì—´í•˜ëŠ” ê²ƒ
    - {'A', 'B', 'C'}ì—ì„œ ì„¸ ê°œë¥¼ ì„ íƒí•˜ì—¬ ë‚˜ì—´í•˜ëŠ” ê²½ìš°: 'ABC', 'ACB', 'BAC', 'BCA', 'CAB', 'CBA'
    
- `ì¡°í•©` : ì„œë¡œ ë‹¤ë¥¸ nê°œì—ì„œ ìˆœì„œì— ìƒê´€ ì—†ì´ ì„œë¡œ ë‹¤ë¥¸ rê°œë¥¼ ì„ íƒí•˜ëŠ” ê²ƒ
    - {'A', 'B', 'C'}ì—ì„œ ìˆœì„œë¥¼ ê³ ë ¤í•˜ì§€ ì•Šê³  ë‘ ê°œë¥¼ ë½‘ëŠ” ê²½ìš° : 'AB', 'AC', 'BC'

```
ìˆœì—´ì˜ ìˆ˜ 
ğ—‡Páµ£ = n * (n - 1) * (n - 2) * ãƒ»ãƒ»ãƒ» * (n - r + 1)
ì¡°í•©ì˜ ìˆ˜
```

````python
# ìˆœì—´
from itertools import permutations

data = ['A', 'B', 'C'] # ë°ì´í„° ì¤€ë¹„

result = list(permutations(data, 3)) # ëª¨ë“  ìˆœì—´ êµ¬í•˜ê¸°
print(result)
````

````python
# ì¡°í•©
from itertools import combinations

data = ['A', 'B', 'C'] # ë°ì´í„° ì¤€ë¹„

result = list(combinations(data, 2)) # 2ê°œë¥¼ ë½‘ëŠ” ëª¨ë“  ì¡°í•© êµ¬í•˜ê¸°
print(result)
````

````python
# ì¤‘ë³µ ìˆœì—´
from itertools import product

data = ['A', 'B', 'C'] # ë°ì´í„° ì¤€ë¹„

result = list(product(data, repeat=2)) # 2ê°œë¥¼ ë½‘ëŠ” ëª¨ë“  ìˆœì—´ êµ¬í•˜ê¸°(ì¤‘ë³µ í—ˆìš©)
print(result)
````

````python
#ì¤‘ë³µ ì¡°í•©
from itertools import combinations_with_replacement

data = ['A', 'B', 'C'] # ë°ì´í„° ì¤€ë¹„

result = list(combinations_with_replacement(data, 2)) # 2ê°œë¥¼ ë½‘ëŠ” ëª¨ë“  ì¡°í•© êµ¬í•˜ê¸°(ì¤‘ë³µ í—ˆìš©)
print(result)
````

## ìµœëŒ€ ê³µì•½ìˆ˜ì™€ ìµœì†Œ ê³µë°°ìˆ˜

- ìµœëŒ€ ê³µì•½ìˆ˜ë¥¼ êµ¬í•´ì•¼ í•  ë•ŒëŠ” `math` ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ `gcd()` í•¨ìˆ˜ë¥¼ ì´ìš©í•  ìˆ˜ ìˆë‹¤.

````python
import math

# ìµœì†Œ ê³µë°°ìˆ˜(LCM)ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜
def lcm(a, b):
    return a * b // math.gcd(a, b)

a = 21
b = 14

print(math.gcd(21, 14)) # ìµœëŒ€ ê³µì•½ìˆ˜(GCD) ê³„ì‚°
print(lcm(21, 14)) # ìµœì†Œ ê³µë°°ìˆ˜(LCM)ê³„ì‚°
````
